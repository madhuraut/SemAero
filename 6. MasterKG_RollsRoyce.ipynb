### Environment Setup

```python
# Importing the required libraries.

import pandas as pd
import numpy as np
```

### Dataset Generation

```python
# Importing the DataFrame for the class "Electric Aircraft".

EA_RollsRoyce = pd.read_csv('EA_RollsRoyce.csv')
EA_RollsRoyce
```

```python
# Importing the DataFrame for the class "Fuel Cell".

FC_RollsRoyce = pd.read_csv('FC_RollsRoyce.csv')
FC_RollsRoyce
```

```python
# Importing the DataFrame for the class "Microgrid".

MG_RollsRoyce = pd.read_csv('MG_RollsRoyce.csv')
MG_RollsRoyce
```

```python
# Importing the DataFrame for the class "Nuclear Reactor".

NR_RollsRoyce = pd.read_csv('NR_RollsRoyce.csv')
NR_RollsRoyce
```

```python
# Combined DataFrame for Master Knowledge Graph.

KG_df = pd.concat([EA_RollsRoyce, FC_RollsRoyce, MG_RollsRoyce, NR_RollsRoyce])
KG_df = KG_df.reset_index(drop=True)
KG_df
```

```python
# Saving the table in a CSV file.

KG_df.to_csv('RollsRoyce_KG.csv', index=False)
```

### Construction of Master Knowledge Graph

```python
# The code is sourced from: https://www.kaggle.com/code/nageshsingh/build-knowledge-graph-using-python.
```

```python
# Importing the required libraries.

import networkx as nx
import matplotlib.pyplot as plt
from tqdm import tqdm
pd.set_option('display.max_colwidth', 200)
%matplotlib inline
```

```python
# Construction of Master Knowledge Graph.

G=nx.from_pandas_edgelist(KG_df, "source", "target", edge_attr=True, create_using=nx.MultiDiGraph())
plt.figure(figsize=(12,12))
pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, node_color='lightgreen', edge_cmap=plt.cm.Greens, pos = pos)
plt.show()
```
